```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate total number of schools
total_schools = school_data_complete["school_name"].nunique()
#print(total_schools)

#Calculate the total number of students
total_students = school_data_complete["student_name"].count()
#print(total_students)

#Calculate the total budget
total_budget = sum(school_data_complete['budget'].unique())
#print(total_budget)

#Calculate the average math score
math_avg = school_data_complete["math_score"].mean()
#print(math_avg)

#Calculate the average reading score
reading_avg = school_data_complete["reading_score"].mean()
#print(reading_avg)

#Calculate the overall passing rate 
overall_passing = (math_avg + reading_avg) / 2
#print(overall_passing)

#Calculate the percentage of students with a passing math score 
math_tote = school_data_complete.loc[school_data_complete["math_score"] >= 70]
math_perc = ((len(math_tote)/total_students) * 100)
#print(math_perc)

#Calculate the percentage of students with a passing reading score
reading_tote = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
reading_perc = ((len(reading_tote)/total_students) * 100)
#print(math_perc)

#Create a dataframe to hold the above results
df_byschool = pd.DataFrame({
    "Total Number of Schools" : [total_schools],
    "Total Number of Students" : [total_students],
    "School Budget" : [total_budget],
    "Math Average" : [math_avg],
    "Reading Average" : [reading_avg],
    "Percent Passing Math" : [math_perc],
    "Percent Passing Reading" : [reading_perc],
    "Overall Passing" : [overall_passing],
})

df_byschool = df_byschool [['Total Number of Schools',
                           'Total Number of Students',
                           'School Budget',
                           'Math Average',
                           'Reading Average',
                           'Percent Passing Math',
                           'Percent Passing Reading',
                           'Overall Passing']]
#Format answers
df_byschool.index.name = None

df_byschool.head()

```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Groupby school name
name_groupby = school_data_complete.groupby(['school_name'])

#Groupyby School Type
sch_types = [x[0] for x in name_groupby['type'].unique()]

#Total Students
total_stus = name_groupby['student_name'].count()
#print(total_stus)

#Total School Budget
total_budg = name_groupby['budget'].unique()
#print(total_budg)

#Per Student Budget
total_stu_budg = total_budg / total_stus
#print(total_stu_budg)

#Average Math score by school
math_sch_avg = name_groupby['math_score'].mean()
#print(math_sch_avg)

#Average Reading score by school
read_sch_avg = name_groupby['reading_score'].mean()
#print(read_sch_avg)

# % Passing Math by school
math_sch_perc = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()/total_stus * 100 
#print(math_sch_perc)

# % Passing Reading by school
read_sch_perc = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/total_stus * 100 
#print(read_sch_perc)

# Overall Passing Rate by school 
overall_sch_pass = (math_sch_perc + read_sch_perc) / 2
#print(overall_passing)

sch_size = school_data_complete['size'].unique()

df_byschool = pd.DataFrame({
    "School Type" : sch_types,
    "Total Students" : total_stus,
    "School Budget" : total_budg,
    "Per Student Budget" : total_stu_budg,
    "Math Average" : math_sch_avg,
    "Reading Average" : read_sch_avg,
    "Percent Passing Math" : math_sch_perc,
    "Percent Passing Reading" : read_sch_perc,
    "Overall Passing" : overall_sch_pass,
    "School Size" : sch_size
})

df_byschool = df_byschool [['School Type',
                            'Total Students',
                            'School Budget',
                            'Per Student Budget',
                            'Math Average',
                            'Reading Average',
                            'Percent Passing Math',
                            'Percent Passing Reading',
                            'Overall Passing',
                           'School Size']]

df_byschool.index.name = "School"

df_byschool.head()

```

## Math Scores by Grade

  * Combine the series into a dataframe

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
grouped_data = df_byschool[["Per Student Budget",
                                     "Math Average",
                                     "Reading Average",
                                     "Percent Passing Math",
                                     "Percent Passing Reading",
                                     "Overall Passing"]]

# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
grouped_data["Spending Ranges"] = pd.cut(df_byschool["Per Student Budget"], spending_bins, labels=group_names)
grouped_data.head()

math_spen = grouped_data.groupby(["Spending Ranges"]).mean()["Math Average"]
read_spen = grouped_data.groupby(["Spending Ranges"]).mean()["Reading Average"]
math_per_spen = grouped_data.groupby(["Spending Ranges"]).mean()["Percent Passing Math"]
read_per_spen = grouped_data.groupby(["Spending Ranges"]).mean()["Percent Passing Reading"]
overall_spen = (math_spen + read_spen)/2

score_by_spending = pd.DataFrame({ 'Math Average' : math_spen,
                                  'Reading Average' : read_spen,
                                  'Percent Passing Math' : math_per_spen,
                                  'Percent Passing Reading' : read_per_spen,
                                  'Overall Passing' : overall_spen
})

score_by_spending = score_by_spending [['Math Average',
                                       'Reading Average',
                                       'Percent Passing Math',
                                       'Percent Passing Reading',
                                       'Overall Passing']]

score_by_spending.head()
```
